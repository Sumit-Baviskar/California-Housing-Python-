#**California Housing Dataset**#

The California Housing Dataset (1990 U.S. census) Describing Various Parameter Related to Location(Longitude And Latitude),Hotel Amentities and Demographical Age,Income.
It contains data related to housing prices in various districts in California.

##**# Understanding the data:**


**Longitude:** distance east or west of the prime meridian

**Latitude:** Latitude lines circle the planet from east and west,
           beginning at 0 at the equator and increasing to 90North or 90South.

**Housing Median Age:** Median Of Age Of People Resident to the
           Housing

**Total Rooms:** Total Number Of Rooms Present.

**Total Bedrooms:**  Total Number Of Rooms Present.

 **Median Income:**Median Income Of Residant In Area  

**Median House Value:** Median Value of House in Area(measured in Dollar($))

**Ocean Proximity:**Proximity To Ocean/Beach

Greenwich Mean Time (GMT) is the mean solar time at the Royal Observatory in Greenwich, London which is measured at midnight. In English speaking nations, GMT is often used synonymously with Universal Time Coordinate (UTC).

#**# Business Question(Data Analysis)**

**1.** What is the **Average Median Income** of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
Grouped_Income= data.groupby(['longitude', 'latitude']) # Group By Function
Average_Income_sales = Grouped_Income.mean() # Finding Mean of Income
print(Average_Income_sales) # Printing the Result
```

```python
plt.figure(figsize=(10,5)) # Ploting the graph Size #width, height
sns.distplot(Average_Income_sales["median_income"], bins=20,color="orange") # Plotinng Distribution Graph
plt.title("AVERAGE INCOME SALES GRAPH")
plt.show() # Printing the Graph
```

As the distribution graph,average mean 3.870 thousand dollars , and range of average income can be seen in 0.40-15.0 thousnad dollar income of the population of california.Most of the population of california region is mode average income (3.3-3.5 thousand dollar) is less than mean average income.

**2.**Show with the help of visualization, how **median_income and median_house_values**  are related?

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
plt.subplot(1,2,1)
a=sns.regplot(x=data['median_income'],y=data['median_house_value'],scatter=False,color="green")
a.set(title="Median Income vs Median House Value ",xlabel="Median Income",ylabel="Median House Value") # To Set Title,X-Label,Y-Label
plt.subplot(1,2,2)

a=sns.histplot(x=data['median_income'],bins=10,color="green",)
a.set(title="Median Income vs Median House Value ",xlabel="Median Income",ylabel="Median House Value") # To Set Title,X-Label,Y-Label
plt.show() # Print the Plot
```

As per both plot, There is a **positive relationship** between medium income and median house value which means as median income increase the value of the house is also increased . Highest value of median house value is 5,00,000 thousnad dollars. Median income of major concentration population is around 2 to 6 thousand dollars income of population.

**3.** Plot **latitude** versus **longitude** and explain your observations.
